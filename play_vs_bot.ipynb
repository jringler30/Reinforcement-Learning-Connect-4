```python
# ── Step 2: Imports ────────────────────────────────────────────────────────
import numpy as np
from tensorflow import keras
from connect4_env import (
    make_board, step, legal_moves, game_over,
    encode_board, print_board, ModelAgent
)

model = keras.models.load_model("m1_pg_final.keras", compile=False)
bot   = ModelAgent(model, sample=False, strong=True)
print("Bot loaded — ready to play!")
```

```python
# ── Step 3: Play a game ────────────────────────────────────────────────────
# Set human_first=True to go first, False to go second.
human_first = True

HUMAN = 1 if human_first else -1
BOT   = -HUMAN

board   = make_board()
current = 1   # player 1 always moves first

print("
Connect-4  |  You = X   Bot = O")
print("Type a column number (0–6) and press Enter to make your move.
")
print_board(board)

while True:
    legal = legal_moves(board)
    done, winner = game_over(board)
    if done:
        break

    if current == HUMAN:
        # Human turn
        while True:
            try:
                col = int(input(f"Your move (legal: {legal}): "))
                if col in legal:
                    break
                print("  Column full or invalid — try again.")
            except ValueError:
                print("  Please enter a number 0–6.")
    else:
        # Bot turn
        col = bot.select_move(board, current)
        print(f"Bot plays column {col}")

    board, _ = step(board, col, current)
    print_board(board)
    current = -current

done, winner = game_over(board)
if winner == HUMAN:
    print("You win!")
elif winner == BOT:
    print("Bot wins!")
else:
    print("Draw!")
```
